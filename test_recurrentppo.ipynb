```python
import gym_examples
import gym
import pygame
from stable_baselines3.common.env_checker import check_env
from stable_baselines3 import PPO, RecurrentPPO 

env = gym.make('gym_examples/GridWorld-v0')
model = RecurrentPPO("MultiInputPolicy", env, verbose=1, device="cpu")
```

```python

model = PPO.load("PPOrecurrent_GridWorld", env=env, device="cpu")
```

```python
model.learn(total_timesteps=300000, reset_num_timesteps=False, tb_log_name="PPO")
episodes = 5
model.save("PPOrecurrent_GridWorld")
```

```python
episodes = 20
for ep in range(episodes):
    obs = env.reset()
    done = False
    while not done:
        action = model.predict(obs)
        #print(type(int(action[0])))
        #print(action[0])
        #print(int(action[0]))
        obs, rewards, done, info, = env.step(action[0])
        env.render()
```

```python
env.close()
```
